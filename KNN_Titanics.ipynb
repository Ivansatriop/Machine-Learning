```python
mean_age = df['Age'].mean()
df = df.fillna(mean_age)
df
```

Preprocessing Data

```python
cat_parch = pd.Series(df['Parch'])
cat_Pclass = pd.Series(df['Pclass'])
cat_SibSp = pd.Series(df['SibSp'])
cat_sex = pd.Series(df['Sex'])

```

```python
cat_parch  = pd.get_dummies(cat_parch)
cat_Pclass = pd.get_dummies(cat_Pclass)
cat_SibSp  = pd.get_dummies(cat_SibSp)
cat_sex = pd.get_dummies(cat_sex)
```

```python
df = df.join(cat_parch)
```

```python
df
```

```python
df = df.rename(columns= {0:'Parch 0', 1:'Parch 1', 2:'Parch 2', 3:'Parch 3', 4:'Parch 4'
                        , 5:'Parch 5', 6:'Parch 6', 9:'Parch 9'})
df.drop(columns='Parch')
```

```python
df = df.join(cat_SibSp)
df = df.rename(columns= {0:'SibSp 0', 1:'SibSp 1', 2:'SibSp 2', 3:'SibSp 3', 4:'SibSp 4'
                        , 5:'SibSp 5', 8:'SibSp 8'})
df
```

```python
df = df.join(cat_Pclass)
df = df.rename(columns= {1:'Pclass 1', 2:'Pclass 2', 3:'Pclass 3'})
df
```

```python
df = df.join(cat_sex)
df
```

```python
df = df.drop(columns=['Pclass','SibSp','Parch', 'Name', 'Sex'])
```

```python
df.info()
```

KNN MODELS

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns='Survived')
y = df.Survived
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)
X_train.shape, X_test.shape
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn.score(X_train, y_train)


```

```python
knn.score(X_test, y_test)
```

```python
y_preds = knn.predict(X_test)
```

USING METRIC SCORE

```python
from sklearn import metrics
```

```python
print('Accuracy:', metrics.accuracy_score(y_test, y_preds))
```
